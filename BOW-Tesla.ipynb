```python
import os
import json
import keras
from keras.preprocessing.text import Tokenizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
import math
import sklearn.metrics as metrics
from sklearn.metrics import confusion_matrix
from pandas_ml import ConfusionMatrix
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
'''
We choose bow as our best model
This model got best accuracy through exp1, exp2 only have +/- labels, not fit this test.

'''
```

```python
num_words=1300


```

```python
text=[]
sentiment=[]
with open('tesla/'+'Tesla_Team2.json','r') as f:
    d = json.load(f)
    text.extend(d['text'].values())
    sentiment.extend(d['sentiment'].values())

X=text
Y=sentiment

print(len(X)==len(Y))
print(len(X))

tokenizer = Tokenizer(num_words=num_words)

tokenizer.fit_on_texts(X)
sequences = tokenizer.texts_to_sequences(X)


X = tokenizer.texts_to_matrix(X)


word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))


(x_train,x_test,y_train,y_test)=train_test_split(X,Y,random_state=42)



ber=BernoulliNB()
ber.fit(x_train,y_train)

y_pred = ber.predict(x_test)


cnf_matrix = ConfusionMatrix(y_test, y_pred)
acc_score = accuracy_score(y_test, y_pred)


print('Accuracy Score: ', acc_score)

print(cnf_matrix)

cnf_matrix.plot()
plt.show()
```
